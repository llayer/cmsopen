```python
def load_samples(inpath = "/eos/user/l/llayer/cmsopen/columnar/note_v0/inferno_systematic/"):
    files = glob.glob(inpath + "*.h5")
    samples = {}
    for sample in files:

        if "cutflow" in sample: continue
        sample_name = sample.split("/")[-1][:-3]
        print(sample_name)
        samples[sample_name] = pd.read_hdf(sample)

    return samples
```

```python
def inferno_to_harvester(file_path, outpath):
    
    f = ROOT.TFile(file_path)
        
    outfile = ROOT.TFile(outpath, 'RECREATE')
    
    # BCE
    outfile.mkdir("bce")
    outfile.cd("bce")    
    h = f.Get("Data" + "_" + "bce_nominal")
    h.Write('data_obs')
    
    for sample in [ "QCD", "TTJets_bkg", "WZJets", "STJets", "TTJets_signal"]:
        h = f.Get(sample + "_" + "bce_nominal")
        h.Write(sample)        
        
        if sample == "TTJets_signal":
            for c in range(4):
                h = f.Get(sample + "_" + 'bce_shift_' + str(c) + '_up')
                h.Write(sample + "_" + 'bce_shift_' + str(c) + 'Up')
                h = f.Get(sample + "_" + 'bce_shift_' + str(c) + '_down')
                h.Write(sample + "_" + 'bce_shift_' + str(c) + 'Down')
                h = f.Get(sample + "_" + 'bce_nuis_' + str(c+1) + '_up')
                h.Write(sample + "_" + 'bce_nuis_' + str(c+1) + 'Up')
                h = f.Get(sample + "_" + 'bce_nuis_' + str(c+1) + '_down')
                h.Write(sample + "_" + 'bce_nuis_' + str(c+1) + 'Down')


    # INFERNO shift
    for c in range(4):
        outfile.mkdir('inferno_shift_' + str(c))
        outfile.cd('inferno_shift_' + str(c)) 
        #"bce_nominal", 'inferno_shift_nominal_' + str(c), 'inferno_nuis_nominal_' + str(c+1)
        h = f.Get("Data" + "_" + 'inferno_shift_nominal_' + str(c))
        h.Write('data_obs')
        
        for sample in [ "QCD", "TTJets_bkg", "WZJets", "STJets", "TTJets_signal"]:
            
            h = f.Get(sample + "_" + 'inferno_shift_nominal_' + str(c))
            h.Write(sample) 
            
            if sample == "TTJets_signal":
                h = f.Get(sample + "_" + 'inferno_shift_' + str(c) + '_up')
                h.Write(sample + "_" + 'inferno_shift_' + str(c) + 'Up')
                h = f.Get(sample + "_" + 'inferno_shift_' + str(c) + '_down')
                h.Write(sample + "_" + 'inferno_shift_' + str(c) + 'Down')
            
    # INFERNO nuis
    for c in range(4):
        outfile.mkdir('inferno_nuis_' + str(c+1))
        outfile.cd('inferno_nuis_' + str(c+1)) 
        h = f.Get("Data" + "_" + 'inferno_nuis_nominal_' + str(c+1))
        h.Write('data_obs')
        
        for sample in [ "QCD", "TTJets_bkg", "WZJets", "STJets", "TTJets_signal"]:

            h = f.Get(sample + "_" + 'inferno_nuis_nominal_' + str(c+1))
            h.Write(sample)         
        
            if sample == "TTJets_signal":
                for i in range(c+1):
                    h = f.Get(sample + "_" + 'inferno_nuis_' + str(c+1) + '_var_' + str(i) + '_up')
                    h.Write(sample + "_" + 'inferno_nuis_' + str(c+1) + '_var_' + str(i) + 'Up')
                    h = f.Get(sample + "_" + 'inferno_nuis_' + str(c+1) + '_var_' + str(i) + '_down')
                    h.Write(sample + "_" + 'inferno_nuis_' + str(c+1) + '_var_' + str(i) + 'Down')
      
        
    

    """
    plot_syst('bce_nominal', 'bce_shift_' + str(c) + '_up',  'bce_shift_' + str(c) + '_down', name='bce_shift_' + str(c))
    plot_syst('bce_nominal', 'bce_nuis_' + str(c+1) + '_up',  'bce_nuis_' + str(c+1) + '_down', name = 'bce_nuis_' + str(c+1))
    plot_syst('inferno_shift_nominal_' + str(c), 'inferno_shift_' + str(c) + '_up',  
              'inferno_shift_' + str(c) + '_down', name = 'inferno_shift_' + str(c))

    for i in range(c+1):
        plot_syst('inferno_nuis_nominal_' + str(c+1), 
                  'inferno_nuis_' + str(c+1) + '_var_' + str(i) + '_up',  
                  'inferno_nuis_' + str(c+1) + '_var_' + str(i) + '_down',
                  name = 'inferno_nuis_' + str(c+1) + '_var_' + str(i)
                 ) 
    """
    
    print(outfile.ls())
    outfile.Close()
        
```
